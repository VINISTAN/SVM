```python
# Import pandas library for accessing csv files
import pandas as pd
```

```python
# Assign dataset to the variable and view the dataset
dataset_1 = pd.read_csv("Social_Network_Ads.csv")
dataset_1.head()
```

```python
dataset_2=pd.get_dummies(dataset_1)
dataset_2
```

```python
# The above dataset i can remove one gender. because i can received the same information
# So avoid space consumptions i can remove one female gender column from the above dataset
```

```python
#Remove female gender column
dataset = pd.get_dummies(dataset_2,drop_first = True)
dataset
```

```python
# Split Input (Independent variable) dataset
independent_variable = dataset[["Gender_Female","Gender_Male","Age","EstimatedSalary"]]
independent_variable
```

```python
# Set output (Dependent variable) dataset
dependent_variable = dataset[["Purchased"]]
dependent_variable
```

```python
# Import Machine Learning Library sklearn and call particular function to split train and test data
from sklearn.model_selection import train_test_split
```

```python
# split train dataset and test dataset using train_test_split function. Assign necessary parameter
x_train,x_test,y_train,y_test = train_test_split(independent_variable,dependent_variable,test_size = 0.25,random_state = 0)
```

```python
# View train and test dataset
x_train,x_test,y_train,y_test
```

```python
# Model Creating
from sklearn.svm import SVC
```

```python
classifier = SVC(C=10.0,kernel="linear",random_state = 0)
```

```python
classifier.fit(x_train,y_train)
```

```python
# Predict the values
y_predict = classifier.predict(x_test)
y_predict
```

```python
# Test the output value of r_score ( r_score >1= Good Model) / ( r_score <1= bad Model)
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_predict)
r_score
```

```python
# Save the Model
import pickle
filename = "purchase_prediction.sav"
pickle.dump(classifier,open(filename,"wb"))
loaded_model = pickle.load(open(filename,"rb"))
```

```python
# Get the input from the user
female = int(input("Are you female? If yes, Enter 1, if No, Enter 0 : "))
male = int(input("Are you male? If yes, Enter 1, if No, Enter 0 : "))
age = int(input("Enter your age : "))
salary = int(input("Enter your salary : "))
result = loaded_model.predict([[female,male,age,salary]])
print("Your decision will be : ",result)
```
